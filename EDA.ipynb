```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Îç∞Ïù¥ÌÑ∞ Î°úÎìú
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

# Îç∞Ïù¥ÌÑ∞ ÌÅ¨Í∏∞ ÌôïÏù∏
print("üîπ Train Îç∞Ïù¥ÌÑ∞ ÌÅ¨Í∏∞:", train.shape)
print("üîπ Test Îç∞Ïù¥ÌÑ∞ ÌÅ¨Í∏∞:", test.shape)

# Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖ ÌôïÏù∏
print("\nüîπ Train Îç∞Ïù¥ÌÑ∞ Ï†ïÎ≥¥:")
print(train.info())

# Í∏∞Î≥∏ ÌÜµÍ≥ÑÎüâ ÌôïÏù∏
print("\nüîπ Train Îç∞Ïù¥ÌÑ∞ Í∏∞Ï¥à ÌÜµÍ≥Ñ:")
print(train.describe())

# Í≤∞Ï∏°Í∞í ÌôïÏù∏
print("\nüîπ Train Îç∞Ïù¥ÌÑ∞ Í≤∞Ï∏°Í∞í Í∞úÏàò:")
print(train.isnull().sum())

print("\nüîπ Test Îç∞Ïù¥ÌÑ∞ Í≤∞Ï∏°Í∞í Í∞úÏàò:")
print(test.isnull().sum())

# ÌóàÏúÑÎß§Î¨º Ïó¨Î∂Ä Î∂ÑÌè¨ ÌôïÏù∏
plt.figure(figsize=(6,4))
sns.countplot(data=train, x='ÌóàÏúÑÎß§Î¨ºÏó¨Î∂Ä', palette="pastel")
plt.title("ÌóàÏúÑÎß§Î¨º Ïó¨Î∂Ä Î∂ÑÌè¨")
plt.show()

# ÏàòÏπòÌòï Î≥ÄÏàò Î∂ÑÌè¨ ÏãúÍ∞ÅÌôî
num_cols = ['Î≥¥Ï¶ùÍ∏à', 'ÏõîÏÑ∏', 'Ï†ÑÏö©Î©¥Ï†Å', 'Ìï¥ÎãπÏ∏µ', 'Ï¥ùÏ∏µ', 'Î∞©Ïàò', 'ÏöïÏã§Ïàò', 'Ï¥ùÏ£ºÏ∞®ÎåÄÏàò', 'Í¥ÄÎ¶¨ÎπÑ']
train[num_cols].hist(figsize=(12, 10), bins=30, layout=(3, 3))
plt.suptitle("ÏàòÏπòÌòï Î≥ÄÏàò Î∂ÑÌè¨")
plt.show()

# Î≤îÏ£ºÌòï Î≥ÄÏàò Í∞í Í∞úÏàò ÌôïÏù∏
cat_cols = ['Îß§Î¨ºÌôïÏù∏Î∞©Ïãù', 'Î∞©Ìñ•', 'Ï£ºÏ∞®Í∞ÄÎä•Ïó¨Î∂Ä', 'Ï§ëÍ∞úÏÇ¨Î¨¥ÏÜå', 'Ï†úÍ≥µÌîåÎû´Ìèº']
for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(data=train, x=col, palette="pastel", order=train[col].value_counts().index)
    plt.title(f"{col} Î∂ÑÌè¨")
    plt.xticks(rotation=45)
    plt.show()

# ÌóàÏúÑÎß§Î¨º Ïó¨Î∂ÄÏóê Îî∞Î•∏ ÏàòÏπòÌòï Î≥ÄÏàò ÎπÑÍµê
plt.figure(figsize=(12, 6))
for i, col in enumerate(num_cols, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(data=train, x='ÌóàÏúÑÎß§Î¨ºÏó¨Î∂Ä', y=col, palette="pastel")
    plt.title(f"{col} Î∂ÑÌè¨")
plt.tight_layout()
plt.show()
```
